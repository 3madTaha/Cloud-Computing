```python
class Student:
    def __init__(self, first, last, email, courses, link):
        self.first = first
        self.last = last
        self.email = email
        self.courses = courses
        self.link = link

    def __str__(self):
        return f"{self.first} {self.last} | {self.email} | {self.courses} | {self.link}"


students = {}

with open('students.txt', 'r') as f:
    for line in f:
        first, last, email, courses, link = line.strip().split(',')
        courses_list = courses.split('|')

        name = f"{first} {last}"
        students[name]=Student(first, last, email, courses_list, link)


names = list(students.keys())

student_name = "Emad Taha" #@param ["Emad Taha"]
favorite = "Breaking Bad" #@param {type:"string"}

student = students[student_name]

print("First Name:", student.first)
print("Last Name:", student.last)
print("Email:", student.email)
print("Courses:", ", ".join(student.courses))
print("Link:", student.link)

button = widgets.Button(
    description="Update",
    button_style='success'
)

output = widgets.Output()

def on_button_clicked(b):
    with output:
        favorite_clean = favorite.strip()

        if not favorite_clean:
            print("Enter favorite first")
            return

        with open('students.txt', 'a') as f:
            f.write(f"\n{student.first},{student.last},{student.email},{'|'.join(student.courses)},{student.link},{favorite_clean}")

        print("Added successfully!")

button.on_click(on_button_clicked)

display(button, output)
```
