```python
from firebase import firebase

FBconn = firebase.FirebaseApplication(
    'https://tutorial3-9522e-default-rtdb.firebaseio.com/',
    None
)
```

```python
def add_word(word):
    data = FBconn.get('/words/', None)

    if data and word in data:
        FBconn.put('/words/', word, data[word] + 1)
    else:
        FBconn.put('/words/', word, 1)
```

```python
def add_text(text):
    words = text.split()
    for word in words:
        add_word(word)
```

```python
def update_word(word, count):
    FBconn.put('/words/', word, count)
```

```python
def delete_word(word):
    FBconn.delete('/words/', word)
```

```python
def show_all():
    data = FBconn.get('/words/', None)
    if data:
        for k, v in data.items():
            print(f"{k}: {v}")
    else:
        print("No data")
```

```python
while True:
    print("\n===== MENU =====")
    print("1. Add word")
    print("2. Add text")
    print("3. Update word")
    print("4. Delete word")
    print("5. Show all")
    print("6. Exit")

    choice = input("Choose: ")

    if choice == "1":
        w = input("Enter word: ")
        add_word(w)

    elif choice == "2":
        t = input("Enter text: ")
        add_text(t)

    elif choice == "3":
        w = input("Word: ")
        while True:
            try:
                c = int(input("New count: "))
                break
            except ValueError:
                print("Invalid input. Please enter a valid integer for the count.")
        update_word(w, c)

    elif choice == "4":
        w = input("Word: ")
        delete_word(w)

    elif choice == "5":
        show_all()

    elif choice == "6":
        print("Bye 👋")
        break

    else:
        print("Invalid choice")
```
